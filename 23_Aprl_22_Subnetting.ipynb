```python
def fonk(start, end) :
    
  start = list(map(int,start.split(".")))
  end = list(map(int,end.split("."))) 

  for i in range(3) :
    end[i+1] += (end[i] - start[i])*256
  return (end[3]-start[3])


# fonk("10.0.0.0", "10.0.1.0")
# fonk("10.0.0.0", "10.0.0.50")
fonk("20.0.0.10", "20.0.1.0")
```

```python

def network_id(ipaddress) :
  ex = ipaddress.split("/")
  ip = ex[0]
  cidr = int(ex[1])
  list_ip = list(map(int,ip.split(".")))
  octet = cidr // 8
  kalan = cidr % 8
  list_ip[octet] = 2**(8-kalan)  * (list_ip[octet] // 2**(8-kalan))
  for i in range(octet+1,4) :
    list_ip[i] = 0
  
  return ".".join(list(map(str,list_ip)))


ip = "200.10.5.68/12"
print(network_id(ip))
```
